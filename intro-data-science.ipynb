# Starting to use the items

```
import pandas as pd
```

```
dataframe = pd.read_csv("/dbfs/databricks-datasets/COVID/CSSEGISandData/csse_covid_19_data/csse_covid_19_daily_reports/04-15-2020.csv")
```

Let's first drop the `FIPS` column and then rename `Admin2` to `City`

```
dataframe = dataframe.drop("FIPS", axis=1)
```

```
dataframe.rename(columns={'Admin2':'City'}, inplace=True)
```

### What can the dataframe tell us?

```
dataframe.sum()
```

```
dataframe.mean()
```

### Lets start manipulating the data
What do we have to play with?

```
dataframe.columns
```

Lets do some basic data manipulation

```
dataframe.sort_values("Confirmed", ascending=False)
```

```
dataframe[dataframe["Country_Region"] == "US"].sort_values("Confirmed", ascending=False)
```

```
dataframe[dataframe["Country_Region"] == "US"].groupby("Province_State")["Confirmed"].sum().sort_values(ascending=False)
```
